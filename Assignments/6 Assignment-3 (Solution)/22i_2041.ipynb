**(b) Generate N-gram Models:**

**i. Develop unigram, bigram, and trigram models for each poet.**

```python
from nltk.util import ngrams
from nltk.probability import FreqDist

unigrams_william = list(ngrams(william, 1))

# Create a list to store valid words
cdf = []

for word in unigrams_william:
    if word[0] not in stop_wordswilliam and all(char not in string.punctuation for char in word[0]):
        cdf.append(word[0])

freq_dist = FreqDist(cdf)
common_wordswilliam = []

for word, frequency in freq_dist.most_common():
  common_wordswilliam.append(word)

for word, frequency in freq_dist.most_common():
    print(f'{word}: {frequency}')

# Print common words
print("Common words:")
print(common_wordswilliam)
```

For each stanza:
For each verse:
Generate a random number in the range [7...10].

```python
import random
import string

choice = input("Enter the style of poetry (1 or 2): ")
print("\n")

if choice == "1":
    for i in range(6):
        # initialize an empty list to create sentences
        poem = []

        # choose a random word from the normalized text of william
        start_word = random.choice(william)
        poem.append(start_word)

        # generate the first part of the poem
        while len(poem) < 15:
            next_words = [word2 for word1, word2 in common_bigrams_william if word1 == start_word]
            if next_words:
                next_word = random.choice(next_words)
                poem.append(next_word)
                start_word = next_word
            else:
                break

        # if the poem is still shorter than 7 words, add random words from the "william" list
        while len(poem) < 15:
            next_word = random.choice(william)
            poem.append(next_word)

        # print the poem without punctuation
        for word in poem:
            if all(char not in string.punctuation for char in word):
                print(word, end=" ")
        print()

        # print an extra newline after the fourth poem
        if i == 3:
            print("\n")

    # generate poem using trigrams
    for i in range(6):
        poem = []
        starting_words = random.choice(william), random.choice(william)
        poem.extend(starting_words)

        while len(poem) < 15:
            next_words = [word3 for (word1, word2, word3) in common_trigrams_william if word1 == starting_words[-2] and word2 == starting_words[-1]]
            if next_words:
                next_word = random.choice(next_words)
                poem.append(next_word)
                starting_words = starting_words[1] + (next_word,)
            else:
                break

        while len(poem) < 15:
            next_word = random.choice(william)
            poem.append(next_word)

        for word in poem:
            if all(char not in string.punctuation for char in word):
                print(word, end=" ")
        print()

        # print an extra newline after the second poem
        if i == 1:
            print("\n")

elif choice == "2":
    for i in range(6):
        verse = []

        # choose a random starting word from the "maya" dataset
        start_maya = random.choice(maya)
        verse.append(start_maya)

        while len(verse) < 15:
            next_words = [word2 for word1, word2 in common_bigrams_maya if word1 == start_maya]
            if next_words:
                next_word = random.choice(next_words)
                verse.append(next_word)
                start_maya = next_word
            else:
                break

        while len(verse) < 15:
            next_word = random.choice(maya)
            verse.append(next_word)

        # print the verse without punctuation
        for word in verse:
            if all(char not in string.punctuation for char in word):
                print(word, end=" ")
        print()

        #print an extra newline after the fourth verse
        if i == 3:
            print("\n")

    for i in range(6):
        verse = []
        starting_words = random.choice(maya), random.choice(maya)
        verse.extend(starting_words)

        while len(verse) < 15:
            next_words = [word3 for (word1, word2, word3) in common_trigrams_maya if word1 == starting_words[-2] and word2 == starting_words[-1]]
            if next_words:
                next_word = random.choice(next_words)
                verse.append(next_word)
                starting_words = starting_words[1] + (next_word,)
            else:
                break

        while len(verse) < 15:
            next_word = random.choice(maya)
            verse.append(next_word)

        # print the verse without punctuation
        for word in verse:
            if all(char not in string.punctuation for char in word):
                print(word, end=" ")
        print()

        # print an extra newline after the second verse
        if i == 1:
            print("\n")
```

You are tasked with solving a puzzle game using Python. Your goal is to reconstruct a complete image from a folder containing random sub-images that are pieces of the original picture.**

**(a) Data Retrieval:
Gather a collection of random sub-images from a specified folder. Each sub image represents a portion of the complete image.**

```python
from PIL import Image
import cv2
import imutils
import numpy as np
import os
from google.colab.patches import cv2_imshow

background = np.ones((600, 900, 3), np.uint8) * 255

# list of image numbers
image_numbers = [1, 5, 18, 7, 3, 6]

for i, img in enumerate(image_numbers):
    image_path = f'img{img}.png'

    if os.path.exists(image_path):
        puzzle = cv2.imread(image_path)

        # check if the image is loaded successfully
        if puzzle is not None:
            # Convert the image to grayscale
            puzzle_gray = cv2.cvtColor(puzzle, cv2.COLOR_BGR2GRAY)

            # resize the grayscale sub-image to match the background size
            puzzle_gray = cv2.resize(puzzle_gray, (background.shape[1] // 3, background.shape[0] // 2))

            if img == 5:
                puzzle_gray = cv2.rotate(puzzle_gray, cv2.ROTATE_90_CLOCKWISE)
            if img == 7:
                puzzle_gray = cv2.flip(puzzle_gray, 1)
            if img == 3:
                puzzle_gray = imutils.rotate(puzzle_gray, angle=315)
            if img == 6:
                # flip and then rotate
                puzzle_gray = cv2.flip(puzzle_gray, 1)
                puzzle_gray = cv2.rotate(puzzle_gray, cv2.ROTATE_90_CLOCKWISE)

            # calculate the position to paste the sub-image
            x1 = i % 3 * (background.shape[1] // 3)
            y1 = i // 3 * (background.shape[0] // 2)

            # convert grayscale to 3 channels
            puzzle_gray = cv2.cvtColor(puzzle_gray, cv2.COLOR_GRAY2BGR)

            # paste the sub-image onto the background
            background[y1:y1 + (background.shape[0] // 2), x1:x1 + (background.shape[1] // 3)] = puzzle_gray

# display the complete image
cv2_imshow(background)
cv2.waitKey(0)
cv2.destroyAllWindows()

# save the complete image
cv2.imwrite('complete_image_gray.jpg', background)
```
